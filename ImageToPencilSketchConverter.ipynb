```python
import cv2
import matplotlib.pyplot as plt
```

```python
 # Read the image
image = cv2.imread("/content/drive/MyDrive/Family.jpg")
```

```python
# Convert the image from BGR to RGB (because OpenCV reads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the original image
plt.figure(figsize=(8, 8))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Display the grayscale image
plt.figure(figsize=(8, 8))
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Invert the grayscale image
inverted_image = 255 - gray_image
```

```python
# Display the inverted grayscale image
plt.figure(figsize=(8, 8))
plt.imshow(inverted_image, cmap="gray")
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

```python
# Apply Gaussian blur to the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
# Invert the blurred image
inverted_blurred = 255 - blurred
```

```python
# Create the pencil sketch by dividing the grayscale image by the inverted blurred image
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

```python
# Display the pencil sketch
plt.figure(figsize=(8, 8))
plt.imshow(pencil_sketch, cmap="gray")
plt.title("Pencil Sketch")
plt.axis("off")
plt.show()
```

```python
# Display the original image and the pencil sketch side by side
plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")


plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch, cmap="gray")
plt.title("Pencil Sketch")
plt.axis("off")

plt.show()
```

Project Summary: Image to Pencil Sketch Converter


This project demonstrates how to convert an image into a pencil sketch using Python and OpenCV. The project involves reading an image, converting it to grayscale, inverting the grayscale image, applying Gaussian blur, and finally, creating a pencil sketch effect by dividing the grayscale image by the inverted blurred image.

Key Steps:
Image Loading: The image is loaded from Google Drive using OpenCV.
Image Conversion: The image is converted from BGR (Blue, Green, Red) to RGB (Red, Green, Blue) for accurate color representation in Matplotlib.
Grayscale Conversion: The original image is converted to grayscale to simplify the color information.

Inversion: The grayscale image is inverted to create a negative of the image.
Blurring: A Gaussian blur is applied to the inverted image to soften it.
Pencil Sketch Creation: The grayscale image is divided by the inverted blurred image to produce the pencil sketch effect.

Visualization: The original image and the pencil sketch are displayed side by side for comparison.


This project is a fun and creative application of basic image processing techniques, demonstrating how to transform images using simple mathematical operations.
